Approach:

Q5. Reverse a number using while loop.

Approach:
- Extract last digit
- Build reversed number

```python
#write answer here
num=1563
reverse=0
while num>0:
  last_digit=num%10
  reverse=reverse*10+last_digit
  num=num//10

print(f"reversed number={reverse}")
```

Q7. Check if a number is palindrome.

Approach:
- Reverse number
- Compare with original

```python
#write answer here
num=15651
original=num
rev=0
while num>0:
  digit=num%10
  rev=rev*10+digit
  num=num//10
if original==rev:
  print("palindrome number")
else:
  print("not a palindrome number")
```
